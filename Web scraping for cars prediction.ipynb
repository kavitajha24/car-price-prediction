```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
```

```python
Path='C:/Users/91865/anaconda3/chromedriver.exe'
```

```python
browser=webdriver.Chrome(executable_path=Path)
```

```python
browser.get('https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001')
```

```python
browser.maximize_window()
```

```python
name=[]
detail=[]
pnd=[]
owner=[]
price=[]
km=[]
mna=[]
emi=[]
```

```python
name1=browser.find_elements_by_xpath("//h2[@class='_3FpCg']")
detail1=browser.find_elements_by_xpath("//p[@class='cvakB']")
#137 to 1943(km . owner, petrol)
km1=browser.find_elements_by_xpath('//ul/li')
owner1=browser.find_elements_by_xpath('//ul/li')
pnd1=browser.find_elements_by_xpath('//ul/li')
emi1=browser.find_elements_by_xpath("//div[@class='_2HFRN']")
price1=browser.find_elements_by_xpath("//div[@class='_7udZZ']")
```

```python
for n in range(len(name1)):
    name.append(name1[n].text)
for d in range(len(detail1)):
    detail.append(detail1[d].text)
```

```python
for e in range(len(emi1)):
    emi.append(emi1[e].text)
for p in range(len(price1)):
    price.append(price1[p].text)
```

```python
for k in range(len(km1)):
    km.append(km1[k].text)
```

```python
for o in range(len(owner1)):
    owner.append(owner1[o].text)
for p in range(len(pnd1)):
    pnd.append(pnd1[p].text)
```

```python
print(len(name))
print(len(detail))
print(len(emi))
print(len(price))
print(len(km))
print(len(pnd))
print(len(owner))
```
